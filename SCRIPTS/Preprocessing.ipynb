```python
# Clean dataset

import pandas as pd

train_df = pd.read_csv('../DATA/Corona_NLP_train.csv', encoding='latin-1')
test_df = pd.read_csv('../DATA/Corona_NLP_test.csv', encoding='latin-1')

df = pd.concat([train_df, test_df], ignore_index=True)

df = df.drop(columns=['UserName', 'ScreenName'])

def strip_unicode(text):
    if not isinstance(text, str):
        return text
    return text.encode('ascii', 'ignore').decode('ascii')

df['OriginalTweet'] = df['OriginalTweet'].apply(strip_unicode)
if 'Location' in df.columns:
    df['Location'] = df['Location'].apply(strip_unicode)

df = df.dropna()

df = df[df['OriginalTweet'].str.strip() != '']

df.to_csv('cleaned_corona_data.csv', index=False)

print(f"Finish Preprocessing, usable data size now: {len(df)}")
print(df.head())
```
